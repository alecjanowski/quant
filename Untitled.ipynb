```python
import math ##For basic mathematical operations
import matplotlib.pyplot as plt
import pylab as pl


from IPython import display ## Plot setup for Ipython
from matplotlib import cm ##  Colormap reference
from matplotlib import gridspec ##plot setups
from matplotlib import pyplot as plt ##plot setups
import numpy as np 
import pandas as pd
from sklearn import metrics
import tensorflow as tf
import matplotlib.patches as mpatches
tf.compat.v1.disable_eager_execution()

df1 = pd.read_csv("./data/AAPL.csv", sep=",")
#df2 = pd.read_csv("./data/AMZN.csv", sep=",")
df3 = pd.read_csv("./data/APRN.csv", sep=",")
df4 = pd.read_csv("./data/BE.csv", sep=",")
df5 = pd.read_csv("./data/CRWD.csv", sep=",")
df6 = pd.read_csv("./data/LOVE.CN.csv", sep=",")
df7 = pd.read_csv("./data/MNS.AX.csv", sep=",")
df8 = pd.read_csv("./data/NFLX.csv", sep=",")
df9 = pd.read_csv("./data/PBYI.csv", sep=",")
df10 = pd.read_csv("./data/SPOT.csv", sep=",")
df11 = pd.read_csv("./data/TSLA.csv", sep=",")
df12 = pd.read_csv("./data/UBER.csv", sep=",")
df13 = pd.read_csv("./data/WORK.csv", sep=",")

#df2 = df2.drop(columns=['Date'])
df3 = df3.drop(columns=['Date'])
df4 = df4.drop(columns=['Date'])
df5 = df5.drop(['Date'], axis=1)
df6 = df6.drop(['Date'], axis=1)
df7 = df7.drop(['Date'], axis=1)
df8 = df8.drop(['Date'], axis=1)
df9 = df9.drop(['Date'], axis=1)
df10 = df10.drop(['Date'], axis=1)
df11 = df11.drop(['Date'], axis=1)
df12 = df12.drop(['Date'], axis=1)
df13 = df13.drop(['Date'], axis=1)


train_x = list(range(len(df1)))


train_y = np.asanyarray(df1[['Open']])




a = tf.Variable(1.0)
b = tf.Variable(160.2)
y = a * train_x + b 

loss = tf.reduce_mean(tf.square(y - train_y))

optimizer = tf.compat.v1.train.GradientDescentOptimizer(0.01)

train = optimizer.minimize(loss)

init = tf.compat.v1.global_variables_initializer()
sess = tf.compat.v1.Session()
sess.run(init)

loss_values = []
train_data = []
for step in range(100):
    _, loss_val, a_val, b_val = sess.run([train, loss, a, b])
    loss_values.append(loss_val)
    
    print(step, loss_val, a_val, b_val)
    train_data.append([a_val, b_val])
        
plt.plot(loss_values, 'ro')

cr, cg, cb = (1.0, 1.0, 0.0)
for f in train_data:
    cb += 1.0 / len(train_data)
    cg -= 1.0 / len(train_data)
    if cb > 1.0: cb = 1.0
    if cg < 0.0: cg = 0.0
    [a, b] = f
    f_y = np.vectorize(lambda x: a*x + b)(train_x)
    line = plt.plot(train_x, f_y)
    plt.setp(line, color=(cr,cg,cb))

plt.plot(train_x, train_y, 'ro')


green_line = mpatches.Patch(color='red', label='Data Points')

plt.legend(handles=[green_line])

plt.show()
```
